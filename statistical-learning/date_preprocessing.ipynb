## Data preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
game_data = pd.read_pickle("../testset.pickle")
game_data_keys = game_data.keys()
print(len(game_data_keys))
```

```python
print(game_data[list(game_data.keys())[0]].keys())
game_data[list(game_data.keys())[0]]
```

```python
df_game_data = pd.DataFrame()
for game_key in game_data_keys:
    # game_key = list(game_data.keys())[0]
    game = game_data[game_key]
    df_game = pd.DataFrame([game])
    # print(df_game.head())
    df_game_data = pd.concat([df_game_data, df_game])
df_game_data.head()
```

```python
df_game_data = df_game_data.reset_index(drop=True)
df_game_data = df_game_data[['attendance', 'start_time', 'day', 'month', 'is_federal_holiday', 
'venue', 'at_night', 'on_grass', 
'temperature', 'wind_speed', 'weather', 'precipitation', 
'team1_name', 'team1_pre_win', 'team1_pre_loss', 
'team1_pre_win_pct', 'team1_players', 'team1_streak', 
'team2_name', 'team2_pre_win', 'team2_pre_loss', 
'team2_pre_win_pct', 'team2_players', 'team2_streak']]
df_game_data.head()
```

```python
df_game_data = df_game_data.reset_index(drop=True)
df_game_data['attendance'] = df_game_data['attendance'].astype(int)
df_game_data_have_aud = df_game_data[df_game_data['attendance'] > 0].reset_index(drop=True)
print(df_game_data_have_aud.shape)
df_game_data_have_aud.head()
```

```python
from datetime import datetime
df_game_data_have_aud['start_hour'] = df_game_data_have_aud['start_time'].apply(lambda x: datetime.strptime(str(x), '%Y-%m-%d %H:%M:%S').hour)
df_game_data_have_aud.drop('start_time', axis=1, inplace=True)
df_game_data_have_aud.head()
```

### deal with players

```python
target_player = pd.read_csv("../data/players/players_all_star_salary_5000000.csv")["Name"].to_numpy()
target_player[0:5]
```

```python
players_5000000 = pd.read_csv("../data/players/players_salary_5000000.csv")
players_8000000 = pd.read_csv("../data/players/players_salary_8000000.csv")
players_15000000 = pd.read_csv("../data/players/players_salary_15000000.csv")
print(players_5000000.shape, players_8000000.shape, players_15000000.shape)
```

```python
df_5000000_8000000 = players_5000000['Name'].isin(players_8000000['Name'])
df_5000000_8000000 = df_5000000_8000000 == False
df_5000000_8000000 = players_5000000[df_5000000_8000000]
df_5000000_8000000 = df_5000000_8000000.reset_index(drop=True)
print(df_5000000_8000000.shape)
```

```python
df_8000000_15000000 = players_8000000['Name'].isin(players_15000000['Name'])
df_8000000_15000000 = df_8000000_15000000 == False
df_8000000_15000000 = players_8000000[df_8000000_15000000]
df_8000000_15000000 = df_8000000_15000000.reset_index(drop=True)
print(df_8000000_15000000.shape)
```

```python
for player in target_player:
    pl = player.replace(" ", "-")
    record = []
    for i in range(len(df_game_data_have_aud)):
        df_player_1 = df_game_data_have_aud['team1_players'][i]
        df_player_2 = df_game_data_have_aud['team2_players'][i]
        if df_player_1["Batting"].str.contains(player).any() or df_player_2["Batting"].str.contains(player).any():
            record.append(1)
        else:
            record.append(0)
    df_game_data_have_aud[pl] = record
```

```python
df_game_data_have_aud.head()
```

```python
rank1 = []
rank2 = []
rank3 = []
for i in range(len(df_game_data_have_aud)):
    df_player_1 = df_game_data_have_aud['team1_players'][i]
    df_player_2 = df_game_data_have_aud['team2_players'][i]
    rank1_num = 0
    rank2_num = 0
    rank3_num = 0

    # calculate 5000000~8000000
    for j in range(len(df_5000000_8000000)):
        if df_player_1["Batting"].str.contains(df_5000000_8000000['Name'][j]).any() or df_player_2["Batting"].str.contains(df_5000000_8000000['Name'][j]).any():
            rank1_num += 1

    # calculate 8000000~15000000
    for j in range(len(df_8000000_15000000)):
        if df_player_1["Batting"].str.contains(df_8000000_15000000['Name'][j]).any() or df_player_2["Batting"].str.contains(df_8000000_15000000['Name'][j]).any():
            rank2_num += 1

    # calculate 15000000~
    for j in range(len(players_15000000)):
        if df_player_1["Batting"].str.contains(players_15000000['Name'][j]).any() or df_player_2["Batting"].str.contains(players_15000000['Name'][j]).any():
            rank3_num += 1

    rank1.append(rank1_num)
    rank2.append(rank2_num)
    rank3.append(rank3_num)
print(len(rank1), len(rank2), len(rank3))
```

```python
df_game_data_have_aud['salary-500-800'] = rank1
df_game_data_have_aud['salary-800-1500'] = rank2
df_game_data_have_aud['salary-1500'] = rank3
```

```python
df_game_data_have_aud.drop(['team1_players', 'team2_players'], axis=1, inplace=True)
```
