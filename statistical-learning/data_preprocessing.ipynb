```python
rank1 = []
rank2 = []
rank3 = []
for i in range(len(df_game_data_have_aud)):
    df_player_1 = df_game_data_have_aud['team1_players'][i]
    df_player_2 = df_game_data_have_aud['team2_players'][i]
    rank1_num = 0
    rank2_num = 0
    rank3_num = 0

    # calculate 5000000~8000000
    for j in range(len(df_5000000_8000000)):
        if df_player_1["Batting"].str.contains(df_5000000_8000000['Name'][j]).any() or df_player_2["Batting"].str.contains(df_5000000_8000000['Name'][j]).any():
            rank1_num += 1

    # calculate 8000000~15000000
    for j in range(len(df_8000000_15000000)):
        if df_player_1["Batting"].str.contains(df_8000000_15000000['Name'][j]).any() or df_player_2["Batting"].str.contains(df_8000000_15000000['Name'][j]).any():
            rank2_num += 1

    # calculate 15000000~
    for j in range(len(players_15000000)):
        if df_player_1["Batting"].str.contains(players_15000000['Name'][j]).any() or df_player_2["Batting"].str.contains(players_15000000['Name'][j]).any():
            rank3_num += 1

    rank1.append(rank1_num)
    rank2.append(rank2_num)
    rank3.append(rank3_num)
print(len(rank1), len(rank2), len(rank3))
```

### Add Column: Average Attendance in previous 5 - 10 games of home team

```python
trainset = pd.read_csv("processed_training.csv")
testset = pd.read_csv("processed_test.csv")

trainset.head()
```

```python
# sort the data by date
trainset = trainset.sort_values(by=['start_time'])
trainset = trainset.reset_index(drop=True)

testset = testset.sort_values(by=['start_time'])
testset = testset.reset_index(drop=True)
```

```python
# for every year, we want to compute the average attendance of previous 5 to 10 games for every home team

# use dictionary to store the attendance for every home team for every game every year
# key: year, value: dictionary
# key: home team, value: every game's attendance
avg_att_in_all_years = {}
for year in range(2014, 2020):
    avg_att_in_this_year = pd.read_csv(f"../data/stadium/avg_attendance_{year}.csv")
    # in avg_att_in_this_year, the first column is the home team name, the second column is the average attendance, we want to transform it to dictionary
    avg_att_in_this_year = avg_att_in_this_year.set_index('Team').T.to_dict('list')
    avg_att_in_all_years[year] = avg_att_in_this_year

teams = list(trainset['team2_name'].unique())
attendance_dict = {}
# create a dataframe to concatenate all the data after adding the moving average attendance
moving_avg_df = pd.DataFrame()
for year in range(2015, 2020):
    attendance_dict[year] = {}
    for team in teams:
        attendance_dict[year][team] = []
        # get the data of the team in the year
        df_team = trainset[(trainset['team2_name'] == team) & (trainset['season'] == year)][['game_page_url', 'team2_name', 'start_time', 'attendance']]
        # create a new column to store the average attendance of previous 5 to 10 games
        df_team['previous_5_to_10MA'] = 0
        df_team.reset_index(drop=True, inplace=True)
        # store the attendance of the team in the year in order to compute the average attendance of previous 5 to 10 games
        for i in range(len(df_team)):
            attendance_dict[year][team].append(df_team['attendance'][i])
            # for the first 10 games, because we don't have previous 5 to 10 games, we use previous year's average attendance for the team
            if i < 10:
                df_team['previous_5_to_10MA'][i] = avg_att_in_all_years[year-1][team][0]
            # for the rest games, we compute the average attendance of previous 5 to 10 games
            else:
                df_team['previous_5_to_10MA'][i] = sum(attendance_dict[year][team][i-10:i-5])/5
        moving_avg_df = pd.concat([moving_avg_df, df_team], axis=0)
        
```

```python
# merge the moving average attendance to the training set
trainset_after_merged = pd.merge(trainset, moving_avg_df[['previous_5_to_10MA', 'game_page_url']], on='game_page_url', how='inner')
trainset_after_merged
```

```python
# for test set, we also want to add a column to store the average attendance of previous 5 to 10 games
attendance_dict_test = {}
test_moving_avg_df = pd.DataFrame()
for team in teams:
    attendance_dict[team] = []
    # get the data of the team in the year
    df_team = testset[testset['team2_name'] == team][['game_page_url', 'team2_name', 'start_time', 'attendance']]
    # create a new column to store the average attendance of previous 5 to 10 games
    df_team['previous_5_to_10MA'] = 0
    df_team.reset_index(drop=True, inplace=True)
    # store the attendance of the team in the year in order to compute the average attendance of previous 5 to 10 games
    for i in range(len(df_team)):
        attendance_dict[team].append(df_team['attendance'][i])
        # for the first 10 games, because we don't have previous 5 to 10 games, we use previous year's average attendance for the team
        if i < 10:
            df_team['previous_5_to_10MA'][i] = avg_att_in_all_years[2019][team][0]
        # for the rest games, we compute the average attendance of previous 5 to 10 games
        else:
            df_team['previous_5_to_10MA'][i] = sum(attendance_dict[2019][team][i-10:i-5])/5
    test_moving_avg_df = pd.concat([test_moving_avg_df, df_team], axis=0)

display(test_moving_avg_df)
```
